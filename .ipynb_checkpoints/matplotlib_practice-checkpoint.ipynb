```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Another way of Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# quick graph Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot. plt.show()---should be used only once per cession
plt.plot(x_axis, y_axis)
plt.show()
```

```python
#create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label = 'Boston')
#create lables for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
#set the y limit between 0 and 45
plt.ylim(0,45)
#create a title
plt.title("PyBer Fare by Month")
#add the legend
plt.legend()
```

```python
#create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue",linewidth=2, label="Boston")
#create lables for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
#set the y limit between 0 and 45
plt.ylim(0,45)
#create a title
plt.title("PyBer Fare by Month")
#add grid
plt.grid()
#add the legend
plt.legend()
```

```python
#skill drill create the plot and add a label for the legend-Object-Oriented Interface
fig,ax=plt.subplots()
ax.plot(x_axis, y_axis,marker="D", color="green",linewidth=2, label="Boston")
#create lables for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
#set the y limit between 0 and 45
ax.set_ylim(0,45)
#create a title
ax.set_title("PyBer Fare by Month")
#add grid
ax.grid()
#add the legend
ax.legend(["Boston"])
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Create the plot
plt.barh(y_axis, x_axis)
```

```python
# Create the plot.----gca()---get current axes chaining with .invert_yaxis() to invert the data set
plt.barh(x_axis, y_axis,color="magenta",label="Boston")
plt.gca().invert_yaxis()
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.legend()
```

```python
# Create the plot with ax.plt()--object oriented interface
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis,color="cyan",label="Chicago")
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.legend()
ax.invert_yaxis()
```

```python
#create scatter plot
plt.plot(x_axis, y_axis, 'o')
```

```python
#skill drill creating scatter plot
plt.scatter(y_axis,x_axis,c="red",label="Chicago")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.legend()
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
plt.xlim(0,45)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
#make the bubble graph larger by using function y_axis_larger, y_axis*3
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#use the list comprehension technique to multiply each data point
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#object-oriented
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
#skill drill
fig, ax = plt.subplots()
ax.scatter(y_axis,x_axis,c="skyblue",edgecolor="black",linewidth=2,s=[i * 5 for i in y_axis],alpha=.8,label="Boston")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.legend()
ax.set_xlim(0,45)
ax.invert_yaxis()
```

```python
#creating pie chart

plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
#exploding the largest data 0-1 
#autopct to format % number
#figsize=() to change the size
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
#skill drill create pie chart - object-oriented interface
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax = plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,labels=x_axis,explode=explode_values,autopct='%.1f%%',shadow="true",startangle=-95,colors=colors)
ax.invert_yaxis()
plt.show()
```
